# **Regressão linear**

### **1. A Equação da Reta (Hipótese)**

Inicialmente, temos a equação da reta. Em aprendizado de máquina, costumamos representá-la com **pesos** e **vieses**. Para uma única variável (uma característica), a equação é:

$$\hat{y}_i = w \cdot x_i + b$$

Onde:
* $\hat{y}_i$ é a **predição do modelo** (o valor que queremos estimar).
* $x_i$ é o **dado de entrada** (a característica, ou *feature*).
* $w$ é o **peso** (*weight*), que define a inclinação da reta.
* $b$ é o **viés** (*bias*), que define onde a reta corta o eixo vertical.

### **2. A Função de Custo (Erro Quadrático Médio)**

Para saber o quão longe a nossa reta está dos dados reais, calculamos a média dos erros ao quadrado. Essa é a função de custo $J$, que queremos minimizar:

$$J(w, b) = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$$

Onde:
* $N$ é o número total de dados observados.
* $y_i$ é o valor real e verdadeiro do dado.
* $\hat{y}_i$ é o valor que a nossa reta previu.


Essa função mede a diferença entre os valores previstos pelo modelo e os valores reais observados. Quanto maior for a diferença entre a previsão do modelo ($\hat{y}_i$) e o valor verdadeiro ($y_i$), maior será o erro ($J$). Ao elevar essas diferenças ao quadrado e somá-las para todos os exemplos, obtemos uma medida global do erro do modelo. O objetivo do treinamento é então **minimizar essa função de custo**, ou seja, encontrar os valores dos parâmetros que produzem o menor erro possível.

### **3. Descida de gradiente**

É nesse ponto que entra o algoritmo de **descida de gradiente**. Esse método é uma técnica de otimização usada para encontrar o mínimo de uma função. Podemos imaginar a função de custo como uma superfície em forma de vale. Cada ponto dessa superfície corresponde a um conjunto específico de valores para os parâmetros do modelo. O valor da função de custo indica a altura nesse terreno. O objetivo do algoritmo é começar em algum ponto inicial e ir descendo gradualmente até chegar ao ponto mais baixo do vale, que representa o menor erro possível.

Para saber em que direção devemos nos mover nessa superfície, usamos o conceito de **derivadas parciais**. Cada parâmetro do modelo possui uma derivada parcial da função de custo associada a ele. Essa derivada indica como o erro muda quando alteramos ligeiramente aquele parâmetro, mantendo os outros constantes. Se a derivada for positiva, significa que aumentar aquele peso aumenta o erro; se for negativa, significa que aumentar o peso reduz o erro. Assim, o algoritmo utiliza essas informações para ajustar os parâmetros sempre na direção que reduz o erro.

### **4. Calculando a derivada parcial com relação ao peso ($w$)**

Primeiro, vamos pegar a nossa função de custo ($J$) e substituir a predição ($\hat{y}_i$) pela equação da reta real, para vermos onde o peso ($w$) está escondido.

Sabemos que a predição é:
$$\hat{y}_i=w \cdot x_i+b$$
Substituindo isso dentro da função de custo $J$, temos:

$$J(w,b)=\frac{1}{N}\sum_{i=1}^{N}(y_i-(w \cdot x_i+b))^2$$

Nós queremos encontrar $\frac{\partial J}{\partial w}$(a derivada parcial da função de custo com relação ao peso), ou seja, como o custo $J$ muda quando mexemos apenas no peso $w$. Para o cálculo dessa derivada, tratamos todas as outras letras ($y_i$, $x_i$, $b$) como se fossem números constantes.

Como temos uma função composta (uma expressão dentro de um parênteses elevado ao quadrado), precisamos usar a **Regra da cadeia**, que diz que a derivada de uma função composta é *a derivada da parte de "fora" (o expoente ao quadrado) multiplicada pela derivada da parte de "dentro" (o conteúdo do parênteses)*.

Para a derivada da parte de "fora", usamos a **Regra do Tombo** (regra básica de derivação), onde o expoente $2$ "cai" multiplicando a base, e o novo exponte passa a ser 1:

$$2 \cdot (y_i-(w \cdot x_i+b))^1$$

Ou, voltando à nossa notação original mais limpa:

$$2 \cdot (y_i-\hat{y}_i)$$

Agora precisamos derivar o conteúdo que estava dentro do parênteses, mas apenas em relação a $w$. O conteúdo é:$$y_i - w \cdot x_i - b$$

Vamos derivar cada termo separadamente em relação a $w$:

* $y_i$: É o valor real, uma constante. A derivada de uma constante é $0$.

* $-b$: É o viés, que neste momento também tratamos como constante. Sua derivada é $0$.

* $-w \cdot x_i$: Aqui está o nosso $w$. Como a derivada de $w$ em relação a ele mesmo é $1$, sobra apenas a constante que o acompanha, que é o $-x_i$.

Portanto, a derivada da parte de "dentro" é simplesmente:

$$-x_i$$

A Regra da Cadeia manda multiplicar esses resultados. Não podemos esquecer o somatório e a fração $\frac{1}{N}$ que estavam lá no começo esperando.

$$\frac{\partial J}{\partial w}=\frac{1}{N}\sum_{i=1}^{N}[2 \cdot (y_i-\hat{y}_i)] \cdot [-x_i]$$

Agora, fazemos uma simples reorganização matemática para deixar a equação mais elegante. Puxamos o $2$ e o sinal de menos ($-$) para a frente da equação:

$$\frac{\partial J}{\partial w}=\frac{-2}{N}\sum_{i=1}^{N}x_i \cdot (y_i-\hat{y}_i)$$

E é exatamente assim que chegamos na fórmula do gradiente para o peso. De modo análogo, a derivada de ($J$) em relação ao viés ($b$) é:

$$\frac{\partial J}{\partial b}=\frac{-2}{N}\sum_{i=1}^{N}(y_i-\hat{y}_i)$$

que é a fórmula do gradiente para o viés ($b$).

### **5. Atualização dos pesos**

A derivada da função de custo em relação ao peso ($w$) serve para medir a inclinação do terreno exatamente no ponto dado.

* **Se a inclinação é positiva** (o gráfico sobe para a direita), significa que o ponto mais baixo da curva está para trás, ou seja, para a **esquerda**.

* **Se a inclinação é negativa** (o gráfico desce para a direita), significa que o ponto mais baixo está para a frente, ou seja, para a **direita**.

Além da direção, a derivada mede a intensidade: um terreno muito íngreme gera um valor de derivada alto, enquanto um terreno quase plano gera um valor próximo a zero.

Com as fórmulas dos gradientes em mãos (que nos diz para onde a função *sobe*), atualizamos os parâmetros na direção contrária (para *descer*). Multiplicamos o gradiente pela **taxa de aprendizado** ($\alpha$), que é o tamanho do passo:

$$w_{novo} = w_{atual} - \alpha \cdot \frac{\partial J}{\partial w}$$

$$b_{novo} = b_{atual} - \alpha \cdot \frac{\partial J}{\partial b}$$

A regra matemática da derivada é que ela aponta para a direção onde a função mais cresce. Como estamos tentando minimizar o erro (ir para baixo), nós precisamos andar na direção contrária. É exatamente por isso que subtraímos:

* Se a derivada é positiva (erro cresce ao aumentar $w$), fazemos $w - (\text{valor positivo})$. O peso atual diminui, movendo-nos para a esquerda, rumo ao fundo do vale.

* Se a derivada é negativa (erro cresce ao diminuir $w$), fazemos $w - (\text{valor negativo})$. Menos com menos dá mais, então o peso atual aumenta, movendo-nos para a direita, também rumo ao fundo do vale.

A derivada sozinha nos fornece informações sobre a direção e a força da inclinação, mas usar esse valor bruto seria perigoso. Se a curva for muito íngreme, o valor da derivada será enorme. Se subtraírmos isso diretamente do peso, teremos um "pulo" gigantesco, podendo passar direto do fundo do vale e ir parar mais alto do outro lado.É aí que entra a multiplicação pelo $\alpha$ (que costuma ser um número pequeno, como $0.01$ ou $0.001$). Ele atua como um regulador:
* Multiplicamos a inclinação pelo $\alpha$ para **escalar o tamanho do passo**. Isso garante andemos na direção certa, mas com um passinho seguro o suficiente para descer suavemente e não ficar "quicando" de um lado para o outro da função de custo.

Assim, em cada iteração do algoritmo, calculamos a derivada da função de custo em relação ao parâmetro e ajustamos seu valor na direção oposta ao crescimento do erro. Esse processo é repetido várias vezes até que os parâmetros se estabilizem e o erro pare de diminuir significativamente.
